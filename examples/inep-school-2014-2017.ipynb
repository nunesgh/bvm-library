# BVM library examples: longitudinal dataset

```python
import pandas
from bvmlib.bvm import BVMLongitudinal
```

## INEP School Census from 2014 to 2017

```python
def load(year):
    source = '/home/nunesgh/temp/inep-school-' + str(year) + '.csv'
    df = pandas.read_csv(source, low_memory=False)
    return df
```

```python
df14 = load(2014)
df15 = load(2015)
df16 = load(2016)
df17 = load(2017)
```

```python
display(df14)
display(df15)
display(df16)
display(df17)
```

```python
df14_ids = ['FK_COD_ALUNO']
df14_qids = ['FK_COD_MUNICIPIO_END','PK_COD_ENTIDADE','FK_COD_ETAPA_ENSINO']
df14_sensitive = ['ID_POSSUI_NEC_ESPECIAL','ID_N_T_E_P']

df15_17_ids = ['CO_PESSOA_FISICA']
df15_17_qids = ['CO_MUNICIPIO_END','CO_ENTIDADE','TP_ETAPA_ENSINO']
```

Description of selected attributes.
- 'FK_COD_MUNICIPIO_END', 'CO_MUNICIPIO_END': city of residency
- 'PK_COD_ENTIDADE', 'CO_ENTIDADE': school id code
- 'FK_COD_ETAPA_ENSINO', 'TP_ETAPA_ENSINO': education level
- 'ID_POSSUI_NEC_ESPECIAL', 'IN_NECESSIDADE_ESPECIAL': disability status
- 'ID_N_T_E_P', 'IN_TRANSPORTE_PUBLICO': public school transportation

### 2014

```python
inep14 = BVMLongitudinal([df14], [df14_ids])
inep14.qids([df14_qids])
inep14.sensitive(df14_sensitive)
```

```python
results14 = inep14.assess()
```

```python
display(results14['re_id'])
display(results14['att_inf'])
```

### 2014 - 2015

```python
inep14_15 = BVMLongitudinal([df14, df15], [df14_ids, df15_17_ids])
inep14_15.qids([df14_qids, df15_17_qids])
inep14_15.sensitive(df14_sensitive)
```

```python
results14_15 = inep14_15.assess()
```

```python
display(results14_15['re_id'])
display(results14_15['att_inf'])
```

### 2014 - 2016

```python
inep14_16 = BVMLongitudinal([df14, df15, df16], [df14_ids, df15_17_ids, df15_17_ids])
inep14_16.qids([df14_qids, df15_17_qids, df15_17_qids])
inep14_16.sensitive(df14_sensitive)
```

```python
results14_16 = inep14_16.assess()
```

```python
display(results14_16['re_id'])
display(results14_16['att_inf'])
```

### 2014 - 2017

```python
inep14_17 = BVMLongitudinal([df14, df15, df16, df17], [df14_ids, df15_17_ids, df15_17_ids, df15_17_ids])
inep14_17.qids([df14_qids, df15_17_qids, df15_17_qids, df15_17_qids])
inep14_17.sensitive(df14_sensitive)
```

```python
results14_17 = inep14_17.assess()
```

```python
display(results14_17['re_id'])
display(results14_17['att_inf'])
```
